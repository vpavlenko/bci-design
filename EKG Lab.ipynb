# Visual analysis of the data

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot(data):
    plt.figure(figsize=(40, 10))
    x = np.linspace(0, len(data), len(data))
    plt.plot(x, data)
    plt.show()
```

We were capturing RA - LA lead EKG data. Here we see QRS complex with 50 Hz noise:

```python
def get_data_from_file(filename):
    return map(int, map(float, open(filename).read().strip().split('\n')))

rest = get_data_from_file('30sec_rest.csv')
plot(rest[:300])
```

# Calculating heart rate variability

```python
from pypeaks import Data, Intervals

data = rest[:500]
data_obj = Data(list(range(len(data))), data, smoothness=2)

data_obj.get_peaks(method='slope')
data_obj.plot()
peaks = data_obj.peaks['peaks'][0]
print(peaks)
```

For simplicity let's assume that we have one tick per 10ms. Let's calculate the variance in peaks:

```python
data = rest
data_obj = Data(list(range(len(data))), data, smoothness=3)

data_obj.get_peaks(method='slope')
data_obj.plot()
peaks = data_obj.peaks['peaks'][0]
print(peaks[:20])
```

```python
pulses = 100. / np.ediff1d(sorted(peaks)) * 60
print(pulses)
```

Here's the answer:

```python
print(np.mean(pulses), np.std(pulses))
```
